```python
import pandas as pd
from pathlib import Path
import time
```

```python
readpath1 = Path.cwd() / 'Resources' / 'cleanData.csv'
readpath2 = Path.cwd() / 'Resources' / 'spotipy.csv'
```

```python
data_df = pd.read_csv(readpath1, encoding='ANSI')
```

```python
spot_df = pd.read_csv(readpath2)
```

```python
data_df.head()
```

```python
tempID = []
for index, row in data_df.iterrows():
    print(f'Working on row {index}')
    try:
        temp = row['URL']
        temp = temp[31:]
        tempID.append(temp)
    except:
        print(f'Error on row {index}')
data_df['TrackID'] = tempID
```

```python
data_df.head()
```

```python
spot_df.head()
```

```python
spot_df.columns
```

```python
spot_df = spot_df.drop(['Unnamed: 0', 'danceability.1'], axis=1)
spot_df.head(100)
```

```python
df = pd.merge(data_df, spot_df, how='left', left_on='TrackID', right_on='trackID')
```

```python
printpath = Path.cwd() / 'Resources' / 'mergedData.csv'
df.to_csv(printpath)
```
